```python
# Dropping rows with missing values
df.dropna(inplace=True)

df.isnull().sum()
```

```python
# Analyzing value frequency in each column
frequency = {}
cols = df.columns
for i in cols:
    frequency[i] = df[i].value_counts()

for i,j in frequency.items():
    print(i,j)
    print()
```

```python
# Selecting numeric columns
columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Histogram for numeric columns
for column in columns:
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Histogram of {column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Frequency')
    plt.show()

```

```python
# KDE Plot for numeric columns
for column in columns:
    sns.kdeplot(data=df[column], cut=5)
    plt.title(f'KDE Plot of {column.capitalize()} with Outliers')
    plt.xlabel(column.capitalize())
    plt.ylabel('Density')
    plt.show()
```

```python
columns=non_numeric_columns = df.select_dtypes(exclude=['object'])

# Create histplot for each column
for column in columns:
    sns.histplot(data=df, x=column)
    plt.title(f'Histogram of {column.capitalize()}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python

# Create histplot for each column
for column in columns:
    sns.boxplot(data=df, y=column)
    plt.title(f'Histogram of {column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.show()
```

```python
for column in columns:
    sns.kdeplot(data=df[column], cut=5)
    plt.title(f'KDE Plot of {column.capitalize()} with Outliers')
    plt.xlabel(column.capitalize())
    plt.ylabel('Density')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Label encoding for categorical variables
categorical_cols = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```
